### Overview

### Import Libraries

```python
import xarray as xr
import numpy as np
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
import os
import requests
import concurrent.futures
%matplotlib inline
```

### Query the CMR API

We will query the GPM granules using its unique Concept ID (C1276812863-GES_DISC), and store the UMM_JSON response to a variable.

```python
# Set link parameters
concept_id = "C1276812863-GES_DISC"
start_time = "2012-03-08T00:00:00Z"
end_time = "2012-03-15T00:00:00Z"

r = requests.get('https://cmr.earthdata.nasa.gov/search/granules.umm_json?collection_concept_id={}&temporal={},{}&pageSize=200'.format(concept_id,
                                                                                                                                        start_time,
                                                                                                                                        end_time))
response_body = r.json()
```

### Parse Cloud OPeNDAP Links from the Response

In UMM_JSON responses from CMR, Cloud OPeNDAP links are located in the <code>RelatedUrls</code> key. Here, we will parse out these OPeNDAP links and store them in a list.

```python
od_files = []
for itm in response_body['items']:
    for urls in itm['umm']['RelatedUrls']:
        if 'OPeNDAP' in urls['Description']:
            od_files.append(urls['URL'])

od_files
        
```

```python
# Print the length of the list
len(od_files)
```

### Add Constraint Expressions for Variable Subsetting, and Download with <code>requests</code>

For Cloud OPeNDAP-enabled granules, [DAP4](https://docs.opendap.org/index.php?title=OPULS_Development#DAP4_Specification) syntax is required for creating constraint expressions if the user wishes to download the granule in a different file format, or for specific variable subsetting.

Below, we create a function called <code>download_granule()</code> that will be called by the ThreadPoolExecutor for downloading the granules in parallel. The function first appends <code>.dap.nc4</code> which is a DAP4 action that will download the granule as a NetCDF4 file. Next, we create our constraint expressions by specifying the variables and slices of those variables we wish to subset.

To explore variables and create links for subsetting, users can access the DAP4 request form (found by appending <code>.dmr.html</code> to the end of the Cloud OPeNDAP granule URL), or visit the [GES DISC Subsetter Tool](disc.gsfc.nasa.gov/information/documents?title=Simple%20Subset%20Wizard%20(SSW)%20Tool%20Retirement%20User%20Guide).

```python
def download_granule(url):
    print (" opening " + url)
    
    # Append ".dap.nc4" to download the granules as NetCDF4 files, with constraint expressions
    data_url = f'{url}.dap.nc4'
    
    """
    required_variables: 
    Variables to be requested as part of the DAP4 Constraint Expression:
    This will request the precipitationQualityIndex variable, including all latitudes, longitudes, 
    and times over the full global extent
    """

    required_variables = {'T2M',
                          'lon',
                          'lat',
                          'time'}

    basename = os.path.basename(data_url)
    request_params = {'dap4.ce': ';'.join(required_variables)}
    response = requests.get(data_url, params=request_params,  headers={'Accept-Encoding': 'identity'})

    # Download the variable if the response is OK
    if response.ok:
        with open(basename, 'wb') as file_handler:
            file_handler.write(response.content)
    else:
        print(f'Request failed: {response.text}')

# Create a pool of 4 workers to download the granules in parallel
with concurrent.futures.ThreadPoolExecutor(max_workers=4) as executor:
    executor.map(download_granule, od_files)


```

### Concatenate and Sum Granules into an Xarray Dataset Object

Next, we can use the <code>open_mfdataset</code> function to concatenate each file into a single Xarray dataset along the 'time' dimension by default.

```python
cloud_data_ds = xr.open_mfdataset('*.dap.nc4')
cloud_data_ds
```

### Calculate Mean 2-meter Temperature

Next, we will use `mean()` to calculate the daily mean temperature, along the time axis. Then, to convert from Kelvin to Celsius, we simply subtract each value by 273.

```python
temp_mean = cloud_data_ds['T2M'].mean(dim="time")
temp_mean -= 273
temp_mean
```

### Plot the Data

Finally, we use Xarray and Cartopy to create a plot of the mean <code>T2M</code> over the Continental United States.

```python
plt.rcParams["figure.figsize"] = (14, 6.5)

# Assuming temp_mean is an xarray DataArray with 'lon' and 'lat' as coordinates
# Create subplot object with projection
ax = plt.subplot(projection=ccrs.PlateCarree())

# Plot temperature using pcolormesh
temp_mean.plot.pcolormesh(x="lon", y="lat", xlim=(-130, -60), ylim=(20, 60), ax=ax)

# Draw the contour line for the freeze line at 0°C and get a reference to the contour object
plt.contour(temp_mean['lon'], temp_mean['lat'], temp_mean, levels=[0], colors='purple', transform=ccrs.PlateCarree())

# Add map features
ax.coastlines()
ax.gridlines(draw_labels=True)
ax.add_feature(cfeature.STATES)

# Create a proxy artist for the contour line (for the legend)
freeze_line = mlines.Line2D([], [], color='purple', label='0°C Mean Freeze Line')

# Create the legend and add the proxy artist
plt.legend(handles=[freeze_line], loc='upper left')

# Add map features
ax.coastlines()
ax.gridlines(draw_labels=True)
ax.add_feature(cfeature.STATES)

# Create title for the figure
plt.suptitle("M2T1NXSLV.5.12.4 Mean Temperature and Freezing Line (°C) from 2012-03-08T00:00:00Z to 2012-03-15T00:00:00Z", fontsize=11.5)

plt.show()
```
